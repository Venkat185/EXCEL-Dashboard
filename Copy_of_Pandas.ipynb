# Merge Data set

```python
students = pd.DataFrame({
    'student_id' : [101,102,103,104],
    'name' : ['Ramesh', 'Suresh', 'Mahesh','Naresh'],
    'city' : ['Hyderabad', 'Vijayawada', 'Guntur', 'Nellore']
})

marks = pd.DataFrame({
    'student_id' : [101, 102, 102, 105],
    'subject' : ['Maths', 'Maths', 'Science', 'Maths'],
    'marks' : [85, 78, 88, 90]
})
```

```python
students.head()
```

```python
marks.head()
```

```python
pd.merge(students, marks, on = "student_id", how = "inner")
```

```python
merge_df = pd.merge(students, marks, on = "student_id", how = "outer")
```

Export data s

```python
merge_df.to_csv("/content/merged_data.csv")
```
